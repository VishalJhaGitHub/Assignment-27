```python
#1. Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

#Ans

#1 - Scatter plot: This is a two-dimensional plot that displays the relationship between two numerical variables. It is useful for identifying trends, clusters, or outliers in the data.

#2 - Line plot: A line plot displays the trend of a numerical variable over time or other continuous variable. It is useful for visualizing changes in the variable over a period of time.

#3 - Histogram: A histogram is a one-dimensional plot that displays the distribution of a numerical variable. It is useful for identifying the range, skewness, and kurtosis of the data.

#4 - Box plot: A box plot displays the distribution of a numerical variable by showing the quartiles, outliers, and median. It is useful for identifying the spread, central tendency, and variability of the data.

#5 - Heatmap: A heatmap displays the relationship between two categorical variables by using color to represent the frequency of occurrence. It is useful for identifying patterns and relationships between two categorical variables.
```

```python
#2. Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x = "timepoint" and y = "signal" for different events and regions.

#Note: timepoint, signal, event, and region are columns in the fmri dataset.

#Ans

import seaborn as sns
import matplotlib.pyplot as plt
# Load the "fmri" dataset
fmri_data = sns.load_dataset("fmri")
# Create a line plot using Seaborn's lineplot() function
sns.lineplot(x="timepoint", y="signal", hue="event", style="region", data=fmri_data)
# Show the plot
sns.set_style("darkgrid")
plt.show()
```

```python
#3. Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x = 'pclass', y = 'age' and y = 'fare'.

#Note: pclass, age, and fare are columns in the titanic dataset.

#Ans

import seaborn as sns
# Load the "titanic" dataset
titanic_data = sns.load_dataset("titanic")
# Create two box plots using Seaborn's boxplot() function
sns.boxplot(x="pclass", y="age", data=titanic_data)
sns.boxplot(x="pclass", y="fare", data=titanic_data)
# Show the plots
sns.set_style("darkgrid")
plt.show()
```

```python
#4. Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue parameter for the 'cut' column of the diamonds dataset.

#Ans

import seaborn as sns
# Load the "diamonds" dataset
diamonds_data = sns.load_dataset("diamonds")
# Create a histogram of the "price" column, colored by the "cut" variable
sns.histplot(data=diamonds_data, x="price", hue="cut")
# Show the plot
sns.set_style("darkgrid")
plt.show()
```

```python
#5. Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column of the iris dataset.

#Ans

import seaborn as sns
# Load the "iris" dataset
iris_data = sns.load_dataset("iris")
# Create a pair plot of all variables, colored by the "species" variable
sns.pairplot(data=iris_data, hue="species")
# Show the plot
sns.set_style("darkgrid")
plt.show()
```

```python
#6. Use the "flights" dataset from seaborn to plot a heatmap.

#Ans

import seaborn as sns
import matplotlib.pyplot as plt
# Load the flights dataset
flights_data = sns.load_dataset("flights")
# Pivot the data to create a matrix of passenger counts by month and year
flights_pivot = flights_data.pivot(index="month", columns="year", values="passengers")
# Create a heatmap of the passenger counts by month and year
sns.heatmap(data=flights_pivot, annot=True, fmt="d", cmap="YlOrRd")
# Set the plot title and axis labels
plt.title("Passenger counts by month and year (2011-2022)")
plt.xlabel("Year")
plt.ylabel("Month")
# Set the x-axis labels to show years from 2011 to 2022
year_labels = [str(year) for year in range(2011, 2023)]
plt.xticks(ticks=[i+0.5 for i in range(len(year_labels))], labels=year_labels, rotation=90)
# Show the plot
plt.show()
```
